```python
#Ans 1:

pip install bokeh

from bokeh.plotting import figure, show

x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

p = figure(title="Simple line plot", x_axis_label='x', y_axis_label='y')

p.line(x, y, legend_label="Line", line_width=2)

p.circle(x, y, legend_label="Points", fill_color="white", size=8)

p.title.text_color = "blue"
p.title.text_font = "times"
p.title.text_font_style = "italic"

show(p)
```

```python
#Ans 2:
'''Bokeh, glyphs are visual representations of data that you can add to plots.
They can include shapes like circles, squares, lines, bars, and more. Adding glyphs to a Bokeh plot 
allows you to visually represent your data in various ways.

Here's how you can add glyphs to a Bokeh plot using an example:
'''

from bokeh.plotting import figure, show

x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

p = figure(title="Glyphs Example", x_axis_label='x', y_axis_label='y')

p.circle(x, y, size=10, color='navy', alpha=0.5)  

p.line(x, y, line_width=2, color='red', alpha=0.8) 

p.square(x, y, size=8, color='green', alpha=0.6) 

show(p)
```

```python
#Ans 3:

from bokeh.plotting import figure, show

# Sample data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a figure object
p = figure(title="Customized Plot", x_axis_label='X Axis', y_axis_label='Y Axis',
           plot_width=600, plot_height=400, background_fill_color='lightgrey',
           outline_line_color='black')

# Add glyphs with legend labels
p.circle(x, y, size=10, color='navy', alpha=0.5, legend_label='Circle')
p.line(x, y, line_width=2, color='red', alpha=0.8, legend_label='Line')

# Customize legend position and appearance
p.legend.location = 'top_left'
p.legend.label_text_font = 'times'
p.legend.label_text_font_style = 'italic'

# Customize axis labels appearance
p.xaxis.axis_label_text_font = 'times'
p.yaxis.axis_label_text_font = 'times'

# Customize title appearance
p.title.text_font = 'times'
p.title.text_font_style = 'bold'
p.title.text_font_size = '16pt'

# Customize grid lines
p.grid.grid_line_color = 'black'
p.grid.grid_line_alpha = 0.3

# Show the plot
show(p)
```

```python
#Ans 5:
'''Create Flask App: Create a Flask application in a Python file (e.g., app.py).'''


from flask import Flask, render_template
from bokeh.embed import server_document

app = Flask(__name__)

@app.route('/')
def index():
    # Replace 'http://localhost:5006/bokeh-app' with the URL of your Bokeh server application
    script = server_document('http://localhost:5006/bokeh-app')
    return render_template('index.html', script=script)

if __name__ == '__main__':
    app.run(debug=True)

'''Create HTML Template: Create an HTML template (e.g., templates/index.html) where the Bokeh plot will be embedded.'''

#html

<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>Bokeh Plot</title>
</head>
<body>
    <h1>Bokeh Plot</h1>
    {{ script|safe }}
</body>
</html>


'''Run Flask App: Run the Flask application.'''


#Using Django:
'''Install Django: If you havent already, install Django using pip:'''

pip install Django

'''Create Django Project: Create a Django project and app.'''


django-admin startproject myproject
cd myproject
python manage.py startapp myapp

'''Define Views: Define views in views.py to serve the Bokeh plot.'''


from django.shortcuts import render
from bokeh.embed import server_document

def bokeh_plot(request):
    # Replace 'http://localhost:5006/bokeh-app' with the URL of your Bokeh server application
    script = server_document('http://localhost:5006/bokeh-app')
    return render(request, 'bokeh_plot.html', {'script': script})

'''Create HTML Template: Create an HTML template (e.g., templates/bokeh_plot.html) where the Bokeh plot will be embedded'''

'''html'''
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>Bokeh Plot</title>
</head>
<body>
    <h1>Bokeh Plot</h1>
    {{ script|safe }}
</body>
</html>
Define URL Configuration: Define URL patterns in urls.py.


from django.urls import path
from . import views

urlpatterns = [
    path('', views.bokeh_plot, name='bokeh_plot'),
]


'''Run Django Server: Run the Django development server.'''


python manage.py runserver
```
