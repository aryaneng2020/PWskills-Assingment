```python
#Ans 1:
import numpy as np
from scipy.stats import pearsonr

# Sample data
study_time = [10, 8, 4, 6, 7]  # hours spent studying
exam_scores = [90, 85, 50, 60, 70]

# Calculate Pearson correlation coefficient
correlation_coefficient, p_value = pearsonr(study_time, exam_scores)

print("Pearson correlation coefficient:", correlation_coefficient)

'''
Interpretation:

-->If the Pearson correlation coefficient is close to 1, it indicates a strong positive linear relationship,
meaning as the amount of time students spend studying increases, their final exam scores tend to increase as well.

-->If the coefficient is close to -1, it indicates a strong negative linear relationship, meaning as the amount of 
time students spend studying increases, their final exam scores tend to decrease.

-->If the coefficient is close to 0, it indicates no linear relationship between the variables.
'''
```

```python
#Ans 2:
from scipy.stats import spearmanr

# Sample data
sleep_hours = [8, 7, 6, 5, 9]  # hours of sleep
job_satisfaction = [9, 7, 5, 4, 8]  # satisfaction level

# Calculate Spearman's rank correlation
correlation_coefficient, p_value = spearmanr(sleep_hours, job_satisfaction)

print("Spearman's rank correlation coefficient:", correlation_coefficient)

'''
Interpretation:

-->If the Spearman's rank correlation coefficient is close to 1, it indicates a strong monotonic increasing relationship,
meaning as the amount of sleep individuals get each night increases, their overall job satisfaction level tends to increase as well.

-->If the coefficient is close to -1, it indicates a strong monotonic decreasing relationship, meaning as the amount of
sleep individuals get each night increases, their overall job satisfaction level tends to decrease.

-->If the coefficient is close to 0, it indicates no monotonic relationship between the variables.
'''
```

```python
#ANs 3:
import numpy as np
from scipy.stats import pearsonr, spearmanr

# Sample data (replace with your actual data)
exercise_hours = [3, 5, 2, 4, 6, ...]  # hours of exercise per week
bmi = [25, 28, 23, 30, 26, ...]  # body mass index

# Calculate Pearson correlation coefficient
pearson_corr_coeff, pearson_p_value = pearsonr(exercise_hours, bmi)

# Calculate Spearman's rank correlation coefficient
spearman_corr_coeff, spearman_p_value = spearmanr(exercise_hours, bmi)

print("Pearson correlation coefficient:", pearson_corr_coeff)
print("Spearman's rank correlation coefficient:", spearman_corr_coeff)

'''
Interpretation:

-->The Pearson correlation coefficient measures the linear relationship between the two variables. 
If it's close to 1, it indicates a strong positive linear relationship, if it's close to -1, it indicates 
a strong negative linear relationship, and if it's close to 0, it indicates no linear relationship.

-->The Spearman's rank correlation coefficient measures the monotonic relationship between the two variables. 
If it's close to 1, it indicates a strong monotonic increasing relationship, if it's close to -1, it indicates 
a strong monotonic decreasing relationship, and if it's close to 0, it indicates no monotonic relationship.

Comparing the results:

-->If both coefficients are similar in magnitude and direction, it suggests that the relationship between exercise hours and BMI 
s both linear and monotonic.

-->If the Pearson coefficient is higher than the Spearman coefficient, it suggests that the relationship is primarily linear.

-->If the Spearman coefficient is higher than the Pearson coefficient, it suggests that the relationship is primarily monotonic
but not necessarily linear.

'''
```

```python
#Ans 5:
import pandas as pd

# Create a DataFrame with the survey results
data = {
    'Age': [25, 42, 37, 19, 31, 28],
    'Soft drink Preference': ['Coke', 'Pepsi', 'Mountain Dew', 'Coke', 'Pepsi', 'Coke']
}

df = pd.DataFrame(data)

# Calculate frequency of each soft drink preference
preference_counts = df['Soft drink Preference'].value_counts()

print("Frequency of each soft drink preference:")
print(preference_counts)

# Calculate average age for each soft drink preference
average_age_by_preference = df.groupby('Soft drink Preference')['Age'].mean()

print("\nAverage age for each soft drink preference:")
print(average_age_by_preference)

```
