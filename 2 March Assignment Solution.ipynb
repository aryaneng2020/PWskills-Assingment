```python
#Ans 2:

'''
Scatter Plot: Displays individual data points as markers, where the position of each point depends on its two-dimensional data coordinates.
'''
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title("Random Data")
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.show()

```

```python
#Ans 3:
'''
The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to arrange multiple plots or axes within the same figure canvas, making it easier to compare different datasets or visualize multiple aspects of your data simultaneously.

Here are some reasons why subplot() is commonly used:

1.Comparison: It enables you to compare different datasets or different views of the same dataset side by side.
2.Visualization of Multiple Aspects: When you need to visualize multiple aspects of your data, such as different variables or different time periods, subplot() helps in organizing these visualizations in a single figure.
3.Compact Presentation: Instead of creating separate figures for each plot, subplot() allows you to present multiple plots within the same figure, resulting in a more compact and coherent visualization.
4.Faceting: Subplots can be used to facet the data, allowing you to break down the data into smaller subsets based on certain categorical variables.
5.Storytelling: It aids in storytelling through data visualization by presenting a series of related plots in a structured and coherent manner.

'''
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3= np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Plot_1')

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Plot_2')

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Plot_3')

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Plot_4')
```

```python
#Ans 4:
'''
A bar plot is a graphical representation of categorical data in which rectangular bars are used to represent the frequencies,
counts, or proportions of each category. The length or height of each bar corresponds to the value being represented.
Bar plots are commonly used to compare the values of different categories or to visualize the distribution of categorical data.

Bar plots are used for several purposes:

1.Comparison of Categorical Data: Bar plots are useful for visually comparing the values of different categories.
For example, you might use a bar plot to compare sales figures for different products or the performance of different teams.

2.Visualization of Counts or Frequencies: Bar plots can be used to represent the counts or frequencies of different categories within a dataset.
This is helpful for understanding the distribution of categorical variables.

3.Identification of Trends: Bar plots can reveal trends or patterns in categorical data. For example, you might observe which categories
have higher or lower values relative to others, indicating trends or patterns in the data.

4.Communication of Results: Bar plots are effective for communicating results to a wider audience, as they provide a clear
and intuitive representation of categorical data. They are commonly used in reports, presentations, and publications.

5.Comparison Across Subgroups: Bar plots can also be used to compare subgroups within each category. By grouping the data into 
subcategories, you can visualize how the values differ across these subgroups.
'''

import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])


plt.bar(company, profit)
plt.title("Company Profit Analysis")
plt.xlabel("Companies")
plt.ylabel("Amount(in CR's)")
plt.show()
```

```python
#Ans 5:
'''
A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset through five summary statistics: 
the minimum, first quartile (Q1), median (second quartile or Q2), third quartile (Q3), and maximum.
It provides a concise way to visualize the central tendency, spread, and skewness of the data, as well as the presence of outliers.

Box plots are used for several purposes:

1.Identifying Central Tendency and Dispersion: Box plots provide a visual summary of the central tendency (median) and dispersion (spread) of the dataset. They allow users to quickly assess the overall shape of the distribution.

2.Comparison of Groups: Box plots are useful for comparing the distributions of different groups or datasets. By placing multiple box plots side by side, users can compare the central tendencies, spreads, and variations across groups.

3.Detection of Outliers: Box plots facilitate the detection of outliers within a dataset. Outliers, which fall beyond the whiskers, are visually distinct, making them easy to identify.

4.Assessment of Symmetry and Skewness: Box plots provide insights into the symmetry or skewness of the data distribution. For instance, if one whisker is longer than the other, it suggests skewness in the data.

5.Robust Visualization: Box plots are robust to extreme values and do not assume a specific underlying distribution of the data, making them suitable for exploring a wide range of datasets.
'''
import matplotlib.pyplot as plt
box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)
plt.subplot(2, 2, 1)
plt.boxplot(box1)
plt.title('Plot_1')
plt.grid(True)

plt.xlabel('Distribution')
plt.ylabel('Values')

plt.subplot(2,2,2)
plt.boxplot(box2)
plt.title('Plot_2')
plt.grid(True)
plt.xlabel('Distribution')
plt.ylabel('Values')
```
