```python
#Ans 10:
import numpy as np
from scipy import stats

# Given data
sample_mean = 6  # average weight loss in pounds
sample_std = 2.5  # standard deviation of weight loss in pounds
sample_size = 50  # number of participants

# Null hypothesis: The drug is not significantly effective (mean weight loss <= 0)
# Alternative hypothesis: The drug is significantly effective (mean weight loss > 0)

# Population parameters under the null hypothesis
population_mean = 0
population_std = sample_std / np.sqrt(sample_size)

# Calculate the t-statistic
t_statistic = (sample_mean - population_mean) / population_std

# Degrees of freedom
df = sample_size - 1

# Calculate the p-value for a one-tailed test
p_value = 1 - stats.t.cdf(t_statistic, df)

# Significance level (alpha)
alpha = 0.05

# Print results
print("Sample Mean:", sample_mean)
print("Sample Standard Deviation:", sample_std)
print("Sample Size:", sample_size)
print("Null Hypothesis: The drug is not significantly effective (mean weight loss <= 0)")
print("Alternative Hypothesis: The drug is significantly effective (mean weight loss > 0)")
print("Degrees of Freedom:", df)
print("T-Statistic:", t_statistic)
print("P-Value:", p_value)

# Compare p-value with significance level to make a decision
if p_value < alpha:
    print("Reject the null hypothesis. The drug is significantly effective.")
else:
    print("Fail to reject the null hypothesis. The drug is not significantly effective.")
```

```python
#Ans 11:
import numpy as np
from scipy.stats import norm

# Given data
sample_proportion = 0.65  # proportion of people satisfied with their job
sample_size = 500  # total number of people surveyed

# Calculate standard error
standard_error = np.sqrt((sample_proportion * (1 - sample_proportion)) / sample_size)

# Calculate z-score for 95% confidence level
z_score = norm.ppf(0.975)  # 95% confidence level is two-tailed

# Calculate margin of error
margin_of_error = z_score * standard_error

# Calculate confidence interval
lower_bound = sample_proportion - margin_of_error
upper_bound = sample_proportion + margin_of_error

# Print results
print("Sample Proportion:", sample_proportion)
print("Sample Size:", sample_size)
print("95% Confidence Interval for the True Proportion of People Satisfied with Their Job:")
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
```

```python
#Ans 12:
import numpy as np
from scipy import stats

# Given data for Sample A
mean_A = 85
std_A = 6

# Given data for Sample B
mean_B = 82
std_B = 5

# Sample sizes are assumed to be equal for simplicity, although you should confirm this in your actual analysis.
# You should also verify whether the variances are equal, as this can affect the choice of the t-test.
sample_size = 30

# Null hypothesis: The two teaching methods have no significant difference in student performance (mean_A - mean_B = 0)
# Alternative hypothesis: The two teaching methods have a significant difference in student performance (mean_A - mean_B ≠ 0)

# Calculate the pooled standard deviation
pooled_std = np.sqrt(((sample_size - 1) * std_A**2 + (sample_size - 1) * std_B**2) / (2 * (sample_size - 1)))

# Calculate the t-statistic
t_statistic = (mean_A - mean_B) / (pooled_std * np.sqrt(2 / sample_size))

# Degrees of freedom
df = 2 * sample_size - 2

# Significance level (alpha)
alpha = 0.01

# Calculate the p-value for a two-tailed test
p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), df))

# Print results
print("Sample A Mean Score:", mean_A)
print("Sample A Standard Deviation:", std_A)
print("Sample B Mean Score:", mean_B)
print("Sample B Standard Deviation:", std_B)
print("Sample Size:", sample_size)
print("Null Hypothesis: The two teaching methods have no significant difference in student performance (mean_A - mean_B = 0)")
print("Alternative Hypothesis: The two teaching methods have a significant difference in student performance (mean_A - mean_B ≠ 0)")
print("Degrees of Freedom:", df)
print("T-Statistic:", t_statistic)
print("P-Value:", p_value)

# Compare p-value with significance level to make a decision
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in student performance between the two teaching methods.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in student performance between the two teaching methods.")
```

```python
#Ans 14:
import numpy as np
from scipy import stats

# Given data
sample_mean = 0.25  # average reaction time in seconds
sample_std = 0.05  # standard deviation of reaction time in seconds
sample_size = 30  # number of participants

# Null hypothesis: Caffeine has no significant effect on reaction time (mu = 0)
# Alternative hypothesis: Caffeine has a significant effect on reaction time (mu ≠ 0)

# Population parameters under the null hypothesis
population_mean = 0

# Calculate the standard error
standard_error = sample_std / np.sqrt(sample_size)

# Calculate the t-statistic
t_statistic = (sample_mean - population_mean) / standard_error

# Degrees of freedom
df = sample_size - 1

# Confidence level
confidence_level = 0.90

# Calculate the critical t-value for a two-tailed test
alpha = 1 - confidence_level
t_critical = stats.t.ppf(1 - alpha / 2, df)

# Calculate the p-value for a two-tailed test
p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), df))

# Print results
print("Sample Mean:", sample_mean)
print("Sample Standard Deviation:", sample_std)
print("Sample Size:", sample_size)
print("Null Hypothesis: Caffeine has no significant effect on reaction time (mu = 0)")
print("Alternative Hypothesis: Caffeine has a significant effect on reaction time (mu ≠ 0)")
print("Degrees of Freedom:", df)
print("T-Statistic:", t_statistic)
print("Critical T-Value:", t_critical)
print("P-Value:", p_value)

# Compare t-statistic with critical t-value and p-value with significance level to make a decision
if abs(t_statistic) > t_critical or p_value < alpha:
    print("Reject the null hypothesis. Caffeine has a significant effect on reaction time.")
else:
    print("Fail to reject the null hypothesis. Caffeine does not have a significant effect on reaction time.")
```
