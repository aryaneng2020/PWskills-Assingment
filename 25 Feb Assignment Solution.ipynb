```python
#ANs 3:
import pandas as pd

course_name = ['Data Science', 'Machine Learning', 'Big Data', 'Data Engineer']
duration =  [2, 3, 6, 4]

df = pd.DataFrame(data={'course_name': course_name, 'duration': duration})

# Define reindex variable
reindex = [3, 0, 1, 2]

# Reindex the DataFrame
new_df = df.reindex(reindex)

# Output for new_df.loc[2]
output_loc = new_df.loc[2]

# Output for new_df.iloc[2]
output_iloc = new_df.iloc[2]

print("Output for new_df.loc[2]:\n", output_loc)
print("\nOutput for new_df.iloc[2]:\n", output_iloc)


'''
new_df.loc[2]: This statement uses label-based indexing to access the row labeled '2'. After reindexing, the row labeled '2' now 
corresponds to the row with index 1 in the original DataFrame (since the reindexing changed the order of the rows).
So, this will return the row with the label '2' in the original DataFrame.

new_df.iloc[2]: This statement uses positional-based indexing to access the third row of the DataFrame (as indexing starts from 0). 
After reindexing, the third row of the DataFrame corresponds to the row with the label '1' in the original DataFrame (since the reindexing changed the order of the rows). So, this will return the third row of the DataFrame after reindexing.

In summary, both statements will return different rows because they use different indexing methods 
(label-based for loc[] and positional-based for iloc[]), and the reindexing operation changed the order of the rows.'''
```

```python
#ANs 5:

import pandas as pd

import numpy as np

columns = ['column_1', 'column_2', 'column_3', 'column_4', 'column_5', 'column_6']
indices = [1,2,3,4,5,6]

#Creating a dataframe:

df1 = pd.DataFrame(np.random.rand(6,6), columns = columns, index = indices)
print(df1)

# Replace the data in the second row of 'column_2' with a string variable
df1.loc[2, 'column_2'] = 'StringData'

# Now, attempting to calculate the mean of 'column_2'
column_2_mean_with_string = df1['column_2'].mean()
print("Mean of column 'column_2' after replacing data with a string variable:", column_2_mean_with_string)


'''
This code will result in an error when attempting to calculate the mean because
the presence of a string value in the column 'column_2' 
will cause Pandas to infer the data type of the column as object (string), 
which cannot be used in numerical operations like mean calculation.
Therefore, replacing numerical data with a string will lead to an error when trying to calculate the mean
'''
```

```python
#Ans 6:
'''
In Pandas, window functions (also known as rolling or moving window functions) are used to perform calculations over a 
specified window of data points. These functions allow you to compute aggregate measures (such as mean, sum, median, etc.)
over a sliding window of data in a DataFrame or Series. Window functions are particularly useful for time-series data analysis,
where you want to analyze trends or patterns within a specific time window.

The types of window functions available in Pandas include:

1.Rolling: This type of window function calculates statistics over a rolling window of a specified size.
You can compute various aggregations such as mean, sum, standard deviation, etc., over this rolling window.

2.Expanding: Expanding window functions calculate statistics from the start of the time series up to the current point.
It expands the window size with each new data point.

3.Exponential Moving Average (EMA): EMA is a type of rolling window function that gives more weight to recent data points
while calculating the average. It's particularly useful for smoothing out data series and reducing noise.

4.Rolling Apply: This function applies a custom function over a rolling window. It allows you to define your own aggregation 
logic for the rolling window.

5.Aggregating: Aggregating window functions calculate summary statistics over a specified window. 
You can use custom aggregation functions to compute metrics such as percentile, skewness, kurtosis, etc.

'''
```

```python
#ANs 8:
import pandas as pd

def calculate_time_difference(start_date, end_date):
    start_datetime = pd.to_datetime(start_date)
    end_datetime = pd.to_datetime(end_date)
    
    time_difference = end_datetime - start_datetime

    A = pd.to_timedelta(time_difference)
   
    return pd.Timedelta(A)

start_date = input("Enter the start date (YYYY-MM-DD): ")
end_date = input("Enter the end date (YYYY-MM-DD): ")

Final = calculate_time_difference(start_date, end_date)

# Display the result
print("Time difference between", start_date, "and", end_date, "is:",Final)

'''
2022-02-13
2022-07-21
'''
```

```python
#Ans 9:
import pandas as pd
import os

def convert_to_categorical(df, column_name, category_order):
    df[column_name] = pd.Categorical(df[column_name], categories=category_order, ordered=True)
    return df

def main():
    file_path = input("Enter the file path: ")
    file_name = os.path.basename(file_path)
    column_name = input("Enter the column name to convert to categorical: ")
    category_order = input("Enter the category order (comma-separated): ").split(',')

    try:
        df = pd.read_csv(file_name)
        df = convert_to_categorical(df, column_name, category_order)
        sorted_df = df.sort_values(by=[column_name])
        
        df1 = pd.DataFrame(sorted_df)
        print(df1)
        df1
    except FileNotFoundError:
        print("File not found.")
    except KeyError:
        print("Column name not found in the CSV file.")
    except Exception as e:
        print("An error occurred:", e)

if __name__ == "__main__":
    main()
    

  
    
#/home/jovyan/work/Assignment/company-sales.csv
#facecream
#comma-separated
```
