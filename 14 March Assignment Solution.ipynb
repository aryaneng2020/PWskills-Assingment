```python
#Ans 3:
import numpy as np
from scipy.stats import f

def variance_ratio_test(data1, data2):
    var1 = np.var(data1, ddof=1)  # unbiased estimate of variance for sample 1
    var2 = np.var(data2, ddof=1)  # unbiased estimate of variance for sample 2

    dfn = len(data1) - 1
    dfd = len(data2) - 1

    f_value = var1 / var2 if var1 > var2 else var2 / var1
    p_value = f.cdf(f_value, dfn, dfd)

    return f_value, dfn, dfd, p_value

# Generate random samples from two normal distributions with known variances
np.random.seed(0)  # for reproducibility
data1 = np.random.normal(loc=0, scale=1, size=100)
data2 = np.random.normal(loc=0, scale=1.5, size=100)

# Perform the variance ratio test
f_value, dfn, dfd, p_value = variance_ratio_test(data1, data2)

# Print the results
print("F-value:", f_value)
print("Degrees of freedom for numerator:", dfn)
print("Degrees of freedom for denominator:", dfd)
print("P-value:", p_value)
```

```python
#Ans 4:
from scipy.stats import f

def f_test(var1, var2, n1, n2):
    dfn = n1 - 1
    dfd = n2 - 1

    f_value = var1 / var2 if var1 > var2 else var2 / var1
    p_value = f.cdf(f_value, dfn, dfd)

    return f_value, dfn, dfd, p_value

# Given data
var1 = 10
var2 = 15
n1 = 12
n2 = 12
alpha = 0.05

# Perform the F-test
f_value, dfn, dfd, p_value = f_test(var1, var2, n1, n2)

# Print the results
print("F-value:", f_value)
print("Degrees of freedom for numerator:", dfn)
print("Degrees of freedom for denominator:", dfd)
print("P-value:", p_value)

# Compare p-value with significance level
if p_value < alpha:
    print("Reject null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject null hypothesis: The variances are not significantly different.")
```

```python
#Ans 5:
from scipy.stats import f

def f_test(var1, var2, n1, n2):
    dfn = n1 - 1
    dfd = n2 - 1

    f_value = var2 / var1  # Since var1 (claimed variance) is in the denominator
    p_value = f.cdf(f_value, dfn, dfd)

    return f_value, dfn, dfd, p_value

# Given data
claimed_variance = 0.005
sample_variance = 0.006
n1 = 1  # Degrees of freedom for the claimed variance (one sample)
n2 = 25  # Sample size for the sample variance
alpha = 0.01

# Perform the F-test
f_value, dfn, dfd, p_value = f_test(claimed_variance, sample_variance, n1, n2)

# Print the results
print("F-value:", f_value)
print("Degrees of freedom for numerator:", dfn)
print("Degrees of freedom for denominator:", dfd)
print("P-value:", p_value)

# Compare p-value with significance level
if p_value < alpha:
    print("Reject null hypothesis: The claim about the variance is not justified.")
else:
    print("Fail to reject null hypothesis: The claim about the variance is justified.")
```

```python
#Ans 6:
def f_distribution_mean_and_variance(dfn, dfd):
    mean = dfd / (dfd - 2) if dfd > 2 else None
    variance = (2 * (dfd ** 2) * (dfn + dfd - 2)) / (dfn * (dfd - 2) ** 2 * (dfd - 4)) if dfd > 4 else None
    return mean, variance

# Example usage:
dfn = 5
dfd = 10
mean, variance = f_distribution_mean_and_variance(dfn, dfd)
print("Mean:", mean)
print("Variance:", variance)
```

```python
#ANs 7:
from scipy.stats import f

# Given data
sample_variance1 = 25
sample_variance2 = 20
n1 = 10
n2 = 15
alpha = 0.10

# Degrees of freedom
dfn = n1 - 1
dfd = n2 - 1

# Calculate the F-value
f_value = sample_variance1 / sample_variance2

# Calculate the critical F-value
critical_f_value = f.ppf(1 - alpha / 2, dfn, dfd)

# Compare the F-value with the critical F-value
if f_value > critical_f_value:
    print("Reject null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject null hypothesis: The variances are not significantly different.")
```

```python
#Ans 8:
from scipy.stats import f

# Given data
waiting_times_restaurant_a = [24, 25, 28, 23, 22, 20, 27]
waiting_times_restaurant_b = [31, 33, 35, 30, 32, 36]
alpha = 0.05

# Calculate the sample variances
sample_variance_a = np.var(waiting_times_restaurant_a, ddof=1)
sample_variance_b = np.var(waiting_times_restaurant_b, ddof=1)

# Degrees of freedom
n1 = len(waiting_times_restaurant_a)
n2 = len(waiting_times_restaurant_b)
dfn = n1 - 1
dfd = n2 - 1

# Calculate the F-value
f_value = sample_variance_a / sample_variance_b

# Calculate the critical F-value
critical_f_value = f.ppf(1 - alpha / 2, dfn, dfd)

# Compare the F-value with the critical F-value
if f_value > critical_f_value:
    print("Reject null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject null hypothesis: The variances are not significantly different.")
```
