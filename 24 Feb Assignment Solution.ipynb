```python
#Ans 2:

import pandas as pd

def reindex_dataframe(df):
  
    new_index = pd.Series(range(1, len(df) * 2, 2))
    df_reindexed = df.reset_index(drop=True).set_index(new_index)
    
    return df_reindexed


data = {'A': [10, 20, 30],
        'B': [40, 50, 60],
        'C': [70, 80, 90]}

df = pd.DataFrame(data)


df_reindexed = reindex_dataframe(df)
print("Old Index:\n",df)
print()
print("New Index:\n",df_reindexed)

```

```python
#Ans 11:
import pandas as pd

def add_weekday_column(df):
    
    df['Date'] = pd.to_datetime(df['Date'])
    
    weekday_names = {0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 4: 'Friday', 5: 'Saturday', 6: 'Sunday'}
    
    df['Weekday'] = df['Date'].apply(lambda x: weekday_names[x.weekday()])
    
    return df


data = {'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05']}
df = pd.DataFrame(data)
df_with_weekday = add_weekday_column(df)
print(df_with_weekday)
```

```python
#Ans 12:
import pandas as pd

def select_rows_within_date_range(df, start_date, end_date):
    
    df['Date'] = pd.to_datetime(df['Date'])
    
    
    selected_rows = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]
    
    return selected_rows


data = {'Date': ['2023-01-01', '2023-01-15', '2023-01-20', '2023-02-01']}
df = pd.DataFrame(data)
start_date = '2023-01-01'
end_date = '2023-01-31'
selected_df = select_rows_within_date_range(df, start_date, end_date)
print(selected_df)
```
